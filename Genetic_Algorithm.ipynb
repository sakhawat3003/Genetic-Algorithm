## *Introduction*

This project presents a simple implementation of a genetic algorithm to illustrate how populations evolve toward optimal solutions. By applying a fitness function and iteratively refining candidates, the algorithm converges to the best-performing solution.

In this notebook, we will walk through each step—population initialization, fitness evaluation, parent selection, crossover, mutation, and replacement. We will see how these processes combine to solve an optimization problem.

This algorithm has the following rules:
1. Find a population of solutions for a certain problem or function
2. Create a fitness function to measure the performance of each solution
3. Select the 2 top most solutions (parents)
4. Represent the performance of each solution by Binary number
5. Random crossover at any point between these two solutions (child)
6. Mutate any binary digit of the child randomly
7. Find the worst performer from the population
8. Replace the worst performer with the new child
9. Repeat the whole process for many generations to get an optimum solution.

```python
import random
```

## *Fitness Function*

```python
def fitness(x):
  return x**2
```

```python
fitness(x=4)
```

## *Integer to Binary and Binary to integer function*

```python
# integer to binary
def int_to_bin(x):
  return format(x, '05b')
```

```python
int_to_bin(5)
```

```python
#binary to integer function
def bin_to_int(b):
  return int(b, 2)
```

```python
bin_to_int('00101')
```

## *Select the Population*

```python
population_size=6
population=[random.randint(0,31) for _ in range(population_size)]
population
```

```python
#find the scores
scores=[fitness(i) for i in population]
scores
```

```python
#create touples for scores with the population so that we can keep track which performace comes from which solution
pairs=list(zip(population, scores))
pairs
```

### *Two top most solutions/Parents*

```python
# find the 2 top most solutions/parents
parents_sorted=sorted(pairs, key=lambda x: x[1], reverse=True)
parents_sorted
```

```python
parent01, parent02= parents_sorted[0][0], parents_sorted[1][0]
print(parent01, parent02)
```

```python
# convert this parents from integer to binary
parent01_bin, parent02_bin=int_to_bin(parent01), int_to_bin(parent02)
print(parent01_bin, parent02_bin)
```

## *Cross between two Parents*

```python
#lets crossover between the parents to produce a child
crossover_point=random.randint(1,4)
crossover_point
```

```python
#binary child after crossover
child_bin=parent01_bin[:crossover_point]+parent02_bin[crossover_point:]
child_bin
```

```python
#integer child
child_int=bin_to_int(child_bin)
child_int
```

```python
#mutation of the child at random point
mutation_point=random.randint(0,4)
mutation_point
```

```python
child_bin_list=list(child_bin)
child_bin_list
```

## *Random Mutation of the Child*

```python
child_bin_list[mutation_point]='1' if child_bin_list[mutation_point]=='0' else '0'
child_bin_list
```

```python
mutated_child=bin_to_int("".join(child_bin_list))
mutated_child
```

## *Replace the Worst Parent/Solution with the Child*

```python
#replace the worst performer from the population with the child
worst_index=scores.index(min(scores))
population[worst_index]=mutated_child
population
```

```python
# best solution based on fitness function
max(population, key=fitness)
```

## *Run the Algorithm for many Generations*

Let's combine all the steps inside a for loop to run the whole process for many generations

```python
for generation in range(10):
  print(f"Generation {generation}: {population}")
  scores=[fitness(i) for i in population]
  pairs=list(zip(population, scores))
  parents_sorted=sorted(pairs, key=lambda x: x[1], reverse=True)
  parent01, parent02= parents_sorted[0][0], parents_sorted[1][0]
  parent01_bin, parent02_bin=int_to_bin(parent01), int_to_bin(parent02)
  crossover_point=random.randint(1,4)
  child_bin=parent01_bin[:crossover_point]+parent02_bin[crossover_point:]
  child_int=bin_to_int(child_bin)
  mutation_point=random.randint(0,4)
  child_bin_list=list(child_bin)
  child_bin_list[mutation_point]='1' if child_bin_list[mutation_point]=='0' else '0'
  mutated_child=bin_to_int("".join(child_bin_list))
  worst_index=scores.index(min(scores))
  population[worst_index]=mutated_child
```

## *Best Solution*

```python
print(f"Final Population: {population}")
print(f"Best Solution: {max(population, key=fitness)}")
```

Indeed for our fitness function 31 produces the biggest output.
